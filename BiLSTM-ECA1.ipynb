```python
import numpy as np
import torch

# ================= GPU检测 =================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


class config():
    data_path = '/新ps_1#_6.8MW.csv'
    timestep = 16
    batch_size = 128
    input_size = 5
    output_size = 1
    num_channels = [16, 32, 64]
    kernel_size = 3
    dropout = 0.01
    epochs = 50
    model_name = 'BiLSTM-ECA1'
    save_path = '/{}.pth'.format(model_name)
    loss_csv_path = '/LOSS/loss_{}.csv'.format(model_name)
    pred_csv_path = '/pred_csv/pred_{}.csv'.format(model_name)
    learning_rate = 0.001


def split_data(data, timestep, feature_size):

    dataX = []
    dataY = []

    for index in range(len(data) - timestep):

        dataX.append(data[index: index + timestep])
        dataY.append(data[index + timestep][4])

    dataX = np.array(dataX)
    dataY = np.array(dataY)

    train_size = int(np.round(0.8 * dataX.shape[0]))

    x_train = dataX[: train_size]
    y_train = dataY[: train_size].reshape(-1, 1)

    x_test = dataX[train_size:]
    y_test = dataY[train_size:].reshape(-1, 1)

    return x_train, y_train, x_test, y_test


import torch.nn as nn


class eca_layer(nn.Module):

    def __init__(self, k_size=3):
        super().__init__()

        self.avg_pool = nn.AdaptiveAvgPool2d(1)

        self.conv = nn.Conv1d(
            1,
            1,
            kernel_size=k_size,
            padding=(k_size - 1) // 2,
            bias=False,
        )

        self.sigmoid = nn.Sigmoid()

    def forward(self, x):

        b, c, t = x.size()

        y = self.avg_pool(x)

        y = self.conv(
            y.squeeze(-1).transpose(-1, -2)
        ).transpose(-1, -2).unsqueeze(-1)

        y = self.sigmoid(y)

        return x * y.expand_as(x)


class BiLSTM_ECA(nn.Module):

    def __init__(
        self,
        input_size,
        output_size,
        num_channels,
        kernel_size,
        dropout,
    ):
        super().__init__()

        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=num_channels[-1],
            batch_first=True,
            bidirectional=True,
        )

        self.eca = eca_layer()

        self.linear = nn.Linear(
            num_channels[-1] * 2,
            output_size,
        )

        self.relu = nn.ReLU()

    def forward(self, x):

        x = x.transpose(1, 2)

        x, _ = self.lstm(x)

        x = x.transpose(1, 2)

        x = self.eca(x)

        x = x[:, :, -1]

        return self.relu(self.linear(x))


from tqdm import tqdm
from sklearn.metrics import *
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import TensorDataset


config = config()

# ================= 读取数据 =================

df = pd.read_csv(config.data_path)

time_col = df["时间"].values
df = df.set_index("时间")
df = df.fillna(0.0)

feature_columns = [
    "平均风向",
    "平均室外温度",
    "平均大气湿度",
    "平均风向对北角度",
    "平均修正风速",
]

df = df[feature_columns]

data = df.values

# ================= 先划分 =================

x_train, y_train, x_test, y_test = split_data(
    data,
    config.timestep,
    config.input_size,
)

# ================= 再归一化 =================

scaler_x = MinMaxScaler()

x_train = scaler_x.fit_transform(
    x_train.reshape(-1, config.input_size)
).reshape(x_train.shape)

x_test = scaler_x.transform(
    x_test.reshape(-1, config.input_size)
).reshape(x_test.shape)


scaler_y = MinMaxScaler()

y_train = scaler_y.fit_transform(y_train)
y_test = scaler_y.transform(y_test)

# ================= tensor =================

x_train = torch.tensor(
    x_train,
    dtype=torch.float32,
).transpose(1, 2)

x_test = torch.tensor(
    x_test,
    dtype=torch.float32,
).transpose(1, 2)

y_train = torch.tensor(y_train, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)


train_loader = torch.utils.data.DataLoader(
    TensorDataset(x_train, y_train),
    config.batch_size,
    False,
)

test_loader = torch.utils.data.DataLoader(
    TensorDataset(x_test, y_test),
    config.batch_size,
    False,
)


model = BiLSTM_ECA(
    config.input_size,
    config.output_size,
    config.num_channels,
    config.kernel_size,
    config.dropout,
).to(device)


loss_function = nn.MSELoss()

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=config.learning_rate,
)


def fit(epoch, model, loss_function, optimizer,
        train_loader, test_loader, bst_loss):

    model.train()
    running_loss = 0

    train_bar = tqdm(train_loader)

    for x_train, y_train in train_bar:

        x_train = x_train.to(device)
        y_train = y_train.to(device)

        optimizer.zero_grad()

        pred = model(x_train)

        loss = loss_function(
            pred,
            y_train.reshape(-1, 1),
        )

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)

    model.eval()

    test_loss = 0

    with torch.no_grad():

        for x_test, y_test in test_loader:

            x_test = x_test.to(device)
            y_test = y_test.to(device)

            pred = model(x_test)

            loss = loss_function(
                pred,
                y_test.reshape(-1, 1),
            )

            test_loss += loss.item()

    test_loss /= len(test_loader)

    if test_loss < bst_loss:

        bst_loss = test_loss

        torch.save(
            model.state_dict(),
            config.save_path,
        )

    print(epoch, epoch_loss, test_loss)

    return epoch_loss, test_loss


train_loss = []
test_loss = []

bst_loss = np.inf

for epoch in range(config.epochs):

    tl, vl = fit(
        epoch,
        model,
        loss_function,
        optimizer,
        train_loader,
        test_loader,
        bst_loss,
    )

    train_loss.append(tl)
    test_loss.append(vl)

print("Finished")
# ===== 保存loss到csv =====

loss_df = pd.DataFrame({
    "epoch": np.arange(1, config.epochs + 1),
    "train_loss": train_loss,
    "val_loss": test_loss,
})

loss_df.to_csv(config.loss_csv_path, index=False)

print("loss saved to:", config.loss_csv_path)

model.eval()

with torch.no_grad():

    pred = model(x_test.to(device)).cpu().numpy()

    true = y_test.numpy()

    pred = scaler_y.inverse_transform(pred)
    true = scaler_y.inverse_transform(true)


model.eval()

with torch.no_grad():

    pred = model(x_test.to(device)).cpu().numpy()

    true = y_test.numpy()

    pred = scaler_y.inverse_transform(pred)
    true = scaler_y.inverse_transform(true)


plt.figure()
plt.plot(true[:200])
plt.plot(pred[:200])
plt.show()


print("R2:", r2_score(true, pred))
print("RMSE:", np.sqrt(mean_squared_error(true, pred)))
print("MAE:", mean_absolute_error(true, pred))
print("MAPE:", mean_absolute_percentage_error(true, pred))


test_time = time_col[-len(true):]

pred_df = pd.DataFrame({
    "时间": test_time,
    "true": true.flatten(),
    "pred": pred.flatten(),
})

pred_df.to_csv(config.pred_csv_path, index=False)

print("pred saved")
```

```python
import pandas as pd

# 1. 读取CSV文件
input_file = r"/pred_BiLSTM-ECA1.csv"
output_file = r"/新pred_BiLSTM-ECA1.csv"

df = pd.read_csv(input_file, encoding='utf-8')

# 2. 将“时间”列转换为datetime格式
df['时间'] = pd.to_datetime(df['时间'])

# 3. 设置时间范围
start_time = pd.to_datetime("2023/8/17 0:00")
end_time = pd.to_datetime("2023/9/16 23:00:00")

# 4. 筛选时间范围内的数据
df = df[(df['时间'] >= start_time) & (df['时间'] <= end_time)]

# ===== 必须重排索引 =====
df = df.reset_index(drop=True)

# ===== 保证能被6整除 =====
df = df.iloc[:len(df)//6*6]

print("筛选后行数:", len(df))


# ================= 每6行合并 =================

group = df.index // 6

agg_dict = {}

for col in df.columns:
    if col == '时间':
        agg_dict[col] = lambda x: x.iloc[0]   # 每组第一行时间
    else:
        agg_dict[col] = "mean"                # 相加/6


df_new = df.groupby(group).agg(agg_dict)

df_new = df_new.reset_index(drop=True)


# 5. 保存
df_new.to_csv(output_file, index=False, encoding='utf-8-sig')

print("数据提取完成，已保存到:", output_file)
print("新数据行数:", len(df_new))
```
